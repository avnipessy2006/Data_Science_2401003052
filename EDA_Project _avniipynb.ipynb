**Handled missing** values in multiple numerical columns by replacing them with their respective mean values. After filling the missing values, I verified that no null values remained in those columns.

* actual_productivity_score

*Library Used*

````python
prod_df["actual_productivity_score"]=prod_df["actual_productivity_score"].fillna(prod_df["actual_productivity_score"].mean())
prod_df["actual_productivity_score"].isnull().sum()
````

````python
prod_df["stress_level"]=prod_df["stress_level"].fillna(prod_df["stress_level"].mean())
prod_df["stress_level"].isnull().sum()
````

````python
prod_df["sleep_hours"]=prod_df["sleep_hours"].fillna(prod_df["sleep_hours"].mean())
prod_df["sleep_hours"].isnull().sum()
````

````python
prod_df["screen_time_before_sleep"]=prod_df["screen_time_before_sleep"].fillna(prod_df["screen_time_before_sleep"].mean())
prod_df["screen_time_before_sleep"].isnull().sum()
````

````python
prod_df["job_satisfaction_score"]=prod_df["job_satisfaction_score"].fillna(prod_df["job_satisfaction_score"].mean())
prod_df["job_satisfaction_score"].isnull().sum()
````

**What I Did**

Created a heatmap to analyze how perceived productivity scores vary across different age groups and social media platform preferences.

***Seaborn***
* heatmap() was used to visualize the data using color intensity.
***Matplotlib***

* Used for figure sizing, labels, and title customization.

**Visualization Interpretation**

* The x-axis represents different social media platforms (Facebook, Instagram, Telegram, TikTok, Twitter).

* Most productivity scores lie between 5.1 and 6.0, indicating relatively small variations across ages and platforms.

**What I Did**

Calculated the average actual productivity score for each gender and displayed the results using a bar chart.

**Libraries Used**

***Pandas***

* groupby() was used to group data by gender.
* mean() was used to calculate the average productivity score.
sort_values() was used to sort the results.
*Matplotlib*
* plot(kind='bar') was used to create the bar chart.
* title(), xlabel(), and ylabel() were used to label the chart.
**Visualization Interpretation**
* The x-axis represents Gender (Female, Male, Other).
* The y-axis represents the Average Actual Productivity Score.
* The average productivity scores for all genders are very close, ranging between approximately 4.8 and 5.0.
* Female participants have the highest average productivity score.
* Male participants have a very similar average score.
* The Other category has a slightly lower average productivity score compared to the other two groups.

**Key Insight**

The chart suggests that gender has little impact on actual productivity scores in this dataset. Productivity levels are relatively consistent across all gender groups, with only minor differences in average scores.

**Conclusion**

There is no significant variation in productivity based on gender. All groups show similar productivity levels, indicating that factors other than gender may play a greater role in influencing productivity.

````python
platform_prod=prod_df.groupby('gender')['actual_productivity_score'].mean().sort_values(ascending=False)
````

````python
platform_prod.plot(kind='bar',)
plt.title("Average Productivity Score by Gender")
plt.xlabel("Gender")
plt.ylabel("AverageProductivity Score")
````
